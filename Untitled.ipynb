```python
import pandas as pd

# Load the dataset
file_path = 'nn_compounds.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
import pandas as pd

# Reload the dataset again with the correct header row
data = pd.read_excel(file_path, header=5)

# Drop any completely empty rows or columns
data.dropna(how='all', inplace=True)
data.dropna(axis=1, how='all', inplace=True)

# Display the cleaned dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Remove columns with more than 50% missing values
threshold = 0.5 * len(data)
data_cleaned = data.dropna(thresh=threshold, axis=1)

# Check the shape of the cleaned data
data_cleaned.shape
```

```python
# Identify categorical and numerical columns
categorical_cols = data_cleaned.select_dtypes(include=['object']).columns
numerical_cols = data_cleaned.select_dtypes(exclude=['object']).columns

# Impute missing values for categorical columns with mode
for col in categorical_cols:
    data_cleaned[col].fillna(data_cleaned[col].mode()[0], inplace=True)

# Impute missing values for numerical columns with median
for col in numerical_cols:
    data_cleaned[col].fillna(data_cleaned[col].median(), inplace=True)

# Verify that there are no missing values left
missing_values_after_imputation = data_cleaned.isnull().sum()
missing_values_after_imputation[missing_values_after_imputation > 0]
```

```python
from sklearn.preprocessing import MinMaxScaler

# Manually identify and convert numerical columns
data_cleaned['Unnamed: 1'] = pd.to_numeric(data_cleaned['Unnamed: 1'], errors='coerce')
data_cleaned['Unnamed: 3'] = pd.to_numeric(data_cleaned['Unnamed: 3'], errors='coerce')
data_cleaned['Unnamed: 4'] = pd.to_numeric(data_cleaned['Unnamed: 4'], errors='coerce')

# Normalize only the numerical columns
scaler = MinMaxScaler()
numerical_cols = ['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 4']
data_cleaned[numerical_cols] = scaler.fit_transform(data_cleaned[numerical_cols])

# Display the first few rows of the normalized data
data_cleaned.head()
```

```python
# Encode categorical variables using one-hot encoding
data_encoded = pd.get_dummies(data_cleaned, columns=categorical_cols)

# List the column names after encoding
data_encoded.columns
```

```python
# Search for columns related to 'Experimental condition' and 'Unnamed: 0'
target_column = [col for col in data_encoded.columns if 'Experimental condition' in col]
identifier_column = [col for col in data_encoded.columns if 'Unnamed: 0' in col]

target_column, identifier_column
```

```python
# Import necessary functions
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Define target columns and drop identifier columns
target_columns = ['Experimental condition_B', 'Experimental condition_G', 'Experimental condition_T']
X_encoded = data_encoded.drop(target_columns + identifier_column, axis=1)
y_encoded = data_encoded[target_columns]

# Split the data into training and test sets
X_train_encoded, X_test_encoded, y_train_encoded, y_test_encoded = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Display the shapes of the training and test sets
X_train_encoded.shape, X_test_encoded.shape, y_train_encoded.shape, y_test_encoded.shape
```

```python
# Decode the one-hot encoded labels
y_train_decoded = y_train_encoded.idxmax(axis=1)
y_test_decoded = y_test_encoded.idxmax(axis=1)

# Initialize and train the model again
model = LogisticRegression(max_iter=1000)
model.fit(X_train_encoded, y_train_decoded)

# Make predictions
y_pred = model.predict(X_test_encoded)

# Evaluate the model
accuracy = accuracy_score(y_test_decoded, y_pred)
report = classification_report(y_test_decoded, y_pred)

# Display the accuracy
print(f'Accuracy Score: {accuracy}\n')

# Display the classification report
print('Classification Report:')
print(report)
```

```python
import data_utils
import os
import numpy as np
import torch
import tqdm
import pathlib
import argparse
import inflect
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize
import model_utils

def get_args():
    parser = argparse.ArgumentParser()
    parser.add_argument('--batch_size', type=int, default=1, help='Batch size for extracting representations')
    parser.add_argument('--models', nargs='+', type=str, default=['phi-3-mini-4k-instruct'], help='Models to use ("phi-3-mini-4k-instruct")')
    parser.add_argument('--layers', nargs='+', type=int, default=None, help='Which layers to use. Defaults to all excluding embedding. First layer indexed by 1')
    parser.add_argument('--device', type=str, default="cpu", help='"cuda" or "cpu"')
    parser.add_argument('--representations', nargs='+', type=str, default=None, help='Representations to extract ("cls", "mean_pooled", "noun_noun_compound")')
    parser.add_argument('--rep_loc', type=str, default="./data", help='Where to save representations')
    parser.add_argument('--save_attention', dest='save_attention', action='store_true', default=False)
    parser.add_argument('--load_if_available', dest='load_if_available', action='store_true', default=False)
    parser.add_argument('--amount_of_dataset', type=float, default=0.05, help='Proportion of dataset to use')

    if not hasattr(__builtins__, 'get_ipython'):
        args = parser.parse_args()
    else:
        args = parser.parse_args(args=[])
        args.batch_size = 1
        args.models = ['phi-3-mini-4k-instruct']
        args.layers = None
        args.device = "cpu"
        args.representations = ["mean_pooled"]
        args.rep_loc = "./data"
        args.save_attention = False
        args.load_if_available = False
        args.amount_of_dataset = 0.05

    return args

def mean_pool_selector(model, model_name, tokeniser, token_reps, input_ids, layer, batch_size, i=None):
    print("model: {model}, model_name: {model_name}, tokeniser: {tokeniser}, token_reps: {token_reps}, layer: {layer}, batch_size:{batch_size}, i: {i}".format(model=model, model_name=model_name, tokeniser=tokeniser, token_reps=token_reps, layer=layer, batch_size=batch_size,i=i))
    non_special_token_mask = lambda x: np.array(tokeniser.get_special_tokens_mask(x, already_has_special_tokens=True)) == 0
    pad_token_mask = lambda x: np.array(x.cpu() == tokeniser.pad_token_id)
    get_tokens_to_keep = lambda x: np.argwhere(non_special_token_mask(x) * (pad_token_mask(x) == False)).reshape(-1)

   
    num_layers = len(token_reps[1]) if isinstance(token_reps, tuple) and len(token_reps) > 1 and model_name in ['phi-3-mini-4k-instruct'] else len(token_reps)

    if layer >= num_layers:
        raise ValueError(f"Layer index {layer} is out of bounds for the model which has {num_layers} layers.")
    
    if isinstance(token_reps, tuple) and len(token_reps) > 1 and model_name in ['phi-3-mini-4k-instruct']:
        layer_reps = token_reps[1][layer].cpu()[:, :, :]  
    elif layer == model.config.num_hidden_layers:
        layer_reps = token_reps[0].cpu()[:, :, :]
    else:
        layer_reps = token_reps[-1][layer].cpu()[:, :, :]

    token_dim = layer_reps.shape[-1]
    print(f"Token dimension: {token_dim}")

    return np.vstack([np.mean(reps[get_tokens_to_keep(input_ids[i])].cpu().numpy(), axis=0) for i, reps in enumerate(layer_reps)])

def get_tokens_from_layers(model_name, model, tokeniser, input_ids, attention_mask, layers, token_selector=mean_pool_selector, load_if_available=True, batch_size=1, rep_type="mean_pooled", torch_device="cuda", save_reps=True, data_loc='./data', add_arg_dict={}, middle_dim=None, save_attention=False):
    rep_locs_per_layer = [data_utils.get_hidden_state_file(model_name, layer=x, rep_type=rep_type, data_loc=data_loc) for x in layers]
    load_reps = load_if_available and os.path.isfile(rep_locs_per_layer[0])
    layer_map = dict(zip(layers, range(len(layers))))
    if load_reps:
        tokens_per_layer = [np.load(x) for x in rep_locs_per_layer]
    else:
        print('Extracting representations from model for layers {}'.format(layers))
        input_ids = input_ids.to(torch_device)
        attention_mask = attention_mask.to(torch_device)
        model.to(torch_device)
        embedding_size = 3072 if 'phi' in model_name else 768
        tokens_per_layer = [np.zeros((input_ids.shape[0], embedding_size)) if middle_dim is None else np.zeros((input_ids.shape[0], middle_dim, embedding_size)) for x in layers]

        with torch.no_grad():
            for i in tqdm.tqdm(range(0, input_ids.shape[0], batch_size)):
                token_reps = model(
                    input_ids[i:i + batch_size, :].reshape(batch_size, -1),
                    attention_mask=attention_mask[i:i + batch_size, :].reshape(batch_size, -1),
                    output_hidden_states=True
                )

                hidden_states = token_reps.hidden_states if hasattr(token_reps, 'hidden_states') else token_reps[1]
                print(f"Layer hidden state shapes: {[t.shape for t in hidden_states]}")

                add_arg_dict["i"] = i

                for layer in layers:
                    if layer < len(hidden_states):
                        tokens_per_layer[layer_map[layer]][i:i + batch_size, :] = token_selector(
                            model, model_name, tokeniser, hidden_states, 
                            input_ids[i:i + batch_size, :].reshape(batch_size, -1), 
                            layer, batch_size, **add_arg_dict
                        )
                    else:
                        print(f"Layer {layer} exceeds available layers. Skipping.")

        if save_reps:
            for layer in layers:
                pathlib.Path('/'.join(rep_locs_per_layer[layer_map[layer]].split('/')[:-1])).mkdir(parents=True, exist_ok=True)
                np.save(rep_locs_per_layer[layer_map[layer]], tokens_per_layer[layer_map[layer]])

    return tokens_per_layer

def extract_and_save_representations(amount_of_dataset=0.05, batch_size=1, models=None, load_if_available=False, layers=None, torch_device="cpu", representations=["mean_pooled"], rep_loc='./data', save_attention=False):
    if representations is None:
        representations = ['cls']

    file_path = 'nn_compounds.xlsx'
    data = data_utils.load_data(file_path)

    sentences = data_utils.get_noun_noun_compound_sentences(data)
    mod_head_words_per_sentence = data_utils.get_noun_noun_mod_head_words_per_sentence(data)

    # Adjust the dataset size
    num_to_keep = int(amount_of_dataset * len(sentences))
    sentences = sentences[:num_to_keep]
    mod_head_words_per_sentence = mod_head_words_per_sentence[:num_to_keep]

    for model_name in tqdm.tqdm(models):
        print(f'Loading {model_name}')
        model, tokeniser = model_utils.load_model(model_name)

        # Dynamically set layers
        if layers is None:
            num_layers = model.config.num_hidden_layers
            print("num_layers is: {}".format(str(num_layers)))
            layers = [min(num_layers - 1, num_layers)]  

        print(f"Extracting representations from model for layers {layers}")

        unpack_dict = lambda x: (x['input_ids'], x['attention_mask'])

        if "mean_pooled" in representations:
            input_ids, attention_mask = unpack_dict(tokeniser.batch_encode_plus(
                sentences,
                max_length=512,
                return_tensors='pt',
                padding=True,  # Padding to max length
                truncation=True  # Truncate to max length
            ))

            
            get_tokens_from_layers(model_name, model, tokeniser, input_ids, attention_mask, layers=layers, load_if_available=load_if_available, batch_size=batch_size, rep_type='mean_pooled', torch_device=torch_device, save_attention=save_attention)


if __name__ == "__main__":
    args = get_args()
    print(vars(args))

    extract_and_save_representations(amount_of_dataset=1, load_if_available=args.load_if_available, batch_size=args.batch_size, models=args.models, layers=args.layers, torch_device=args.device, representations=args.representations, rep_loc=args.rep_loc, save_attention=args.save_attention)
```
